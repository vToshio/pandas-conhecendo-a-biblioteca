## Atividade 2


```python
import pandas as pd

url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'

dados = pd.read_csv(url, sep=';')
```

```python
df = dados.query('Tipo == "Apartamento"')
df.head()
```

```python
# 1 - Média de Quartos por Apartamento
df.Quartos.mean()
```

```python
# 2 - Quantos Bairros únicos existem
len(df.Bairro.unique())
```

```python
# ouuuuu:
df.Bairro.nunique()
```

```python
# 3 - Quais bairros possuem a média de valor mais elevada

df_bairro_valor = df.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False).head()
df_bairro_valor
```

```python
df_bairro_valor.plot(
    kind='barh',
    figsize=(14,10),
    color='red'
)
```

## Atividade 3

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
df = pd.read_csv(url, sep=',')
```

### Verificar a existência de valores nulos e tratá-los

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
df
```

### Remover "Alice" e "Carlos"

```python
query = df.query('Nome in ["Alice", "Carlos"]').index
```

```python
df.drop(query, axis=0, inplace=True)
df
```

### Aplique um filtro que exiba apenas os alunos que foram aprovados

```python
# Primeira forma
alunos_aprovados = df.query('Aprovado == True')
alunos_aprovados
```

```python
# Segunda forma
filtro = (df['Aprovado'] == True)
alunos_aprovados = df[filtro]
alunos_aprovados
```

### Salvar os Alunos Aprovados em um arquivo CSV

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False)
pd.read_csv('alunos_aprovados.csv')
```

### Extra:
Ao conferir as notas dos alunos aprovados, notamos que algumas notas estavam incorretas. As alunas que tiraram nota 7.0, na verdade, tinham um ponto extra que não foi contabilizado. Sendo assim, substitua as notas 7.0 da base de dados por 8.0. 

**Dica:** pesquise pelo método replace.

```python
alunos_aprovados.replace(
    to_replace = 7.0,
    value = 8.0,
)
```

```python
alunos_aprovados.to_csv('alunos_aprovados_correcao.csv', index=False)
pd.read_csv('alunos_aprovados_correcao.csv')
```

```python
alunos_aprovados.plot(kind='bar', x='Nome', y='Notas')
```
