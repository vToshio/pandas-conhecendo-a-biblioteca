#### Agrupando dados de acordo com o tipo de imóvel
- Utilização do método .groupby(): objetiva dividir os dados em grupos, para aplicar alguma operação neles
   - Semelhante à lógica do SQL
   - Parâmetros:
      - by (str): especificar a coluna que queremos agrupas
      - axis (int): especificar o eixo ao longo do qual queremos agrupas
         - 0: linhas; 1: colunas
      - sort (booleano): indica se os grupos devem ser ordenados pelo valor da coluna de agrupamento
      - dropna (booleano): controlar se valores ausentes serão excluídos ou não     
- Obs: é adequado realizar uma distinção entre dados de imóveis comerciais e imóveis residenciais, pois pode vir a prejudicar o aprendizado do modelo

````python
# Funciona sempre em conjunto com outros tipos de métodos do pandas
# como: mean, sum, min...
dados.groupby('Tipo').mean(numeric_only=True) #vai ser aplicado em todas as colunas de valore numéricos
````

````python
# Exibe os dados como específicos da coluna Valor em um DataFrame, e ordenado pela própria coluna de Valor
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
````

````python
# Armazenando em uma variável
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

# Plotando essas informações em um gráfico
df_preco_tipo.plot(
    kind='barh', # Tipo do Gráfico
    figsize = (14, 10), # Largura x Altura do gráfico
    color = 'purple' # Cor do gráfico
)
````

## Tratando Dados Nulos
- Não tratar dados pode interferir diretamente na qualidade das análises e resultados obtidos
   - Pode ocorrer **problemas** como: **viés**, **redução de acurácia** e **imprecisão nas informações obtidas**
   - Algumas análises e modelos estátisticos não são capazes de lidar com a presença de dados faltando
- Assegura a **confiabilidade** e a **precisão** das **análises e decisões tomadas** a partir dos dados

### Como posso realizar o tratamento de dados nulos?
- **.dropna()**: remove todas as linhas ou colunas que possuem pelo menos um único valor nulo
- **.fillna()**: preenche todos os valores nulos por um valor específico (value)
- **.interpolate()**: realiza uma interpolação dos dados, preenchendo os valores nulos com valores interpolados (a partir dos vizinhos)
  

````python
# Retorna um data frame booleano que mostra os campos que são nulos(True) ou não (False)
df.isnull()
````

````python
# Retorna uma series com a quantidade de dados nulos em cada tabela
df.isnull().sum()
````

````python
# Substitui os campos nulos por algum valor (definido pelo escopo do projeto)
df = df.fillna(value=0)
df
````

````python
df.isnull().sum()
````

````python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
````

````python
dados.head()
````

### 1. Valor por mês
Essa coluna deve conter os gastos mensais de cada imóvel, incluindo aluguel e condomínio

````python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
````

### 2. Valor por ano
Essa coluna deve conter os gastos anuais por imóvel (IPTU + 12 X (ALUGUEL + CONDOMÍNIO))

````python
dados['Valor_por_ano'] = dados['IPTU'] + 12 * (dados['Valor_por_mes'])
````

````python
dados
````

## Criar Colunas Categóricas
- **Descrição:** contêm uma sumarização das principais informações dos imóveis (tipo do imóvel, bairro, quantidade de quartos e vagas)
- **Possui_suite:** informe apenas se o imóvel possui ou não suítes, sem que importar com a quantidade

````python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                     dados['Quartos'].astype(str) + \
                    ' quarto(s) e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
````

````python
# Utilização do método .apply(), que aplica uma função para cada item de um
# iterável em uma coluna, no caso, da coluna 'Suites'
dados['Possui_suite'] = dados['Suites'].apply(lambda x: 'Sim' if x>0 else 'Não')
dados.head()
````
